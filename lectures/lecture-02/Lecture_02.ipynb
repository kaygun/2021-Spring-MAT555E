```python
G = nx.random_regular_graph(3,20)

length = nx.shortest_path_length(G, source=0)

nodelist,hops = zip(*length.items())

positions = nx.drawing.nx_agraph.graphviz_layout(G, prog='twopi', root=0)

nx.draw(G, positions, nodelist = nodelist, node_color=hops, cmap=plt.cm.copper)
plt.axis('equal')
plt.savefig('graph.png')
```
