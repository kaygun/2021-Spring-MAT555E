```python
from random import randint

import numpy as np
import pandas as pd
import scipy.stats

import matplotlib.pyplot as plt
```

Here is a simple question. Assume I have a set of input output pairs given below

```python
xs = np.random.uniform(0,1,100)
xs.sort()
ys = 0.01*xs**2 + 2.1*xs + 1.3 + np.random.normal(0,0.2,100)
```

If I construct the model below, would there be a statistically significant difference between the outputs and model?

```python
model1 = 2*xs + 1.4 + np.random.normal(0,0.2,100)

plt.plot(xs,ys)
plt.plot(xs,model1)
```

How do we test something like this?

## Paired t-test

[Paired t-test](https://en.wikipedia.org/wiki/Student%27s_t-test) is a statistical test is suitable for testing whether any two models that take numeric values as inputs and yields numeric outputs on the same input dataset do show any statistically significant differences.

Let us see if we can see that:

```python
scipy.stats.ttest_rel(ys,model1)
```

Read upon paired t-test. The result above says the test statistics is non-zero, but is statistically **insignificant**. They may differ, but the difference is insignificant.

On the other hand, if we form another model and test:

```python
model2 = 2*xs + 1.45 + np.random.normal(0,0.1,100)

plt.plot(xs,ys)
plt.plot(xs,model2)

scipy.stats.ttest_rel(ys,model2)
```

The result above shows these models are different and difference is statistically significant. 

### What did we just test?

The paired t-test tests whether the difference between the models on each point on average are significant.

This is not the only way we can test whether two models $f,g\colon \mathbb{R}\to \mathbb{R}$ are different.

## Mean Square Error, Mean Absolute Error

Assume we have a set of inputs $(x_i | i=1,\ldots,n)$ linked with outputs $(y_i)$ and a set of model outputs $(\hat{y}_i)$. The mean square error and the mean absolute error between the outputs and the model is calculated as

$$ MSE = \frac{\sum_i (y_i - \hat{y}_i)^2}{n} \qquad MAE = \frac{\sum_i |y_i - \hat{y}_i|}{n} $$
